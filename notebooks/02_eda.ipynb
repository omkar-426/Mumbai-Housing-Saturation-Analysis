## Key EDA Findings

- Mumbai HPI grew from ~90 to ~420 between 2010 and 2025 — a 4.6x increase
- Growth was highest in 2011-2014 (25-32% YoY), driven by post-2008 recovery and infrastructure boom
- A clear deceleration is visible post-2016, with growth dropping below 10%
- The 2020-2022 plateau is partially a COVID-19 artefact — lockdowns suppressed transactions and construction, not true market saturation
- Post-2022, growth resumed but remains low (1-5% YoY), suggesting the market is genuinely approaching saturation
- The overall HPI trend visually resembles a logistic (S-curve) growth pattern — suitable for saturation modelling

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("D:\Mumbai-Housing-Saturation\data\processed\mumbai_hpi_clean.csv")
print(df.head())
print(df.shape)
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df['time'], df['hpi'], marker='o', markersize=3, linewidth=1.5, color='steelblue')
plt.title('Mumbai Housing Price Index (2010-2025)')
plt.xlabel('Year')
plt.ylabel('HPI (Base: 2010-11 = 100)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('../outputs/figures/mumbai_hpi_trend.png', dpi=150)
plt.show()
```

```python
df['yoy_growth'] = df['hpi'].pct_change(periods=4) * 100

plt.figure(figsize=(12, 4))
plt.bar(df['time'], df['yoy_growth'], color='steelblue', alpha=0.7, width=0.2)
plt.axhline(0, color='red', linewidth=0.8)
plt.title('Mumbai HPI - Year-on-Year Growth Rate (%)')
plt.xlabel('Year')
plt.ylabel('YoY Growth (%)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('../outputs/figures/mumbai_hpi_yoy_growth.png', dpi=150)
plt.show()
```
