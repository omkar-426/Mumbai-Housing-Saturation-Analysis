# 01 - Data Cleaning

**Source:** Reserve Bank of India - Housing Price Index (City-wise, Quarterly)  
**File:** `data/raw/House Price Index Publication.xlsx`

## What this notebook does
- Loads the raw RBI HPI Excel file containing 3 tables across different base years
- Extracts Table 2 (Base Year: 2010-11 = 100) for Mumbai only
- Covers Q1 2010-11 to Q4 2024-25 (76 quarterly observations)
- Converts quarter labels (e.g. Q1.2010-11) to decimal year format (e.g. 2010.0)
- Saves cleaned data to `data/processed/mumbai_hpi_clean.csv`

## Output columns
- `quarter` — original quarter label (e.g. Q1.2010-11)
- `hpi` — Mumbai Housing Price Index value
- `time` — decimal year (used for modelling)

```python
import pandas as pd
import openpyxl
```

```python
wb = openpyxl.load_workbook("D:/Mumbai-Housing-Saturation/data/raw/House Price Index Publication.xlsx", read_only=True)
ws = wb.active

for i, row in enumerate(ws.iter_rows(values_only=True)):
    print(i, row)
    if i > 80:
        break
```

```python
for i, row in enumerate(ws.iter_rows(values_only=True)):
    if 24 <= i <= 35:
        print(i, row)
        
```

```python
rows = []
for i, row in enumerate(ws.iter_rows(values_only=True)):
    if i >= 28 and i <= 87 and row[1] is not None:
        quarter = str(row[1]).replace('(P)**', '').strip()
        mumbai = row[10]
        if isinstance(quarter, str) and quarter.startswith('Q') and mumbai is not None:
            rows.append({'quarter': quarter, 'hpi': mumbai})

df = pd.DataFrame(rows)
print(df.tail())
print(df.shape)
```

```python
def quarter_to_decimal(q):
    parts = q.split('.')
    qnum = int(parts[0][1])
    year = int(parts[1].split('-')[0])
    return year + (qnum - 1) * 0.25

df['time'] = df['quarter'].apply(quarter_to_decimal)
print(df.head(8))
```

```python
df.to_csv("../data/processed/mumbai_hpi_clean.csv", index=False)
print("Saved!")
```
